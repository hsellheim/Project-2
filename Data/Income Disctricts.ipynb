```python
import pandas as pd
df = pd.read_excel('Income districts.xlsx')
```

```python
df = df.drop(0)
df = df.drop(1)
df = df.drop(2)
df = df.drop(4)
```

```python
df = df.rename(columns={"Unnamed: 2": "County ID", "Unnamed: 7": "County", "Unnamed: 30": "Income 2022"})
```

```python
df = df.iloc[406:421].reset_index(drop=True)
```

```python
df = df[["County ID", "County", "Income 2022"]]
```

```python
df["Income 2022"] = df["Income 2022"] * 1_000_000
```

```python
df_pop = pd.read_excel('Population Counties.xlsx')
```

```python
df_pop.head()
```

```python
df_pop = df_pop.rename(columns={"Unnamed: 7": "County", "Unnamed: 35": "Population 2022"})
```

```python
df_pop = df_pop[["County", "Population 2022"]]
```

```python
df_merged = df.merge(df_pop, on='County')
```

```python
df_merged["Income per capita 2022"] = df_merged["Income 2022"] / df_merged["Population 2022"]
```
